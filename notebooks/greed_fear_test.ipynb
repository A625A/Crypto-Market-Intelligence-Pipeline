```python
"""
Fetch Fear & Greed Index data from Alternative.me API.
"""

import json
import logging
from pathlib import Path

import requests

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)

logger = logging.getLogger(__name__)
```

```python
RAW_PATH = Path("data/raw/fear_greed_raw.json")


def fetch_fear_greed(limit=0):
    url = "https://api.alternative.me/fng/"

    params = {
        "limit": limit,
        "format": "json"
    }

    logger.info("Fetching Fear & Greed data from Alternative.me")

    response = requests.get(url, params=params, timeout=10)
    response.raise_for_status()

    data = response.json()

    logger.info("Fetched %s Fear & Greed records", len(data.get("data", [])))

    return data

print (f"Got {len(data)} records")
```
